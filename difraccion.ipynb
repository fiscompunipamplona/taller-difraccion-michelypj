Funcion de Bessel para m=1

```python
from math import*
from numpy import*
tramos = 1000
a = 0
b = pi
h = (b-a)/tramos

def function_x_theta(tramos,a,h):
    xn1 = 0
    xn2 = 0
    m = 1
    for i in range(0,21,1):
        for j in range(1,int((tramos/2)+1)):
            theta1 = a +(2*j-1)*h
            y1 = (h/(3*pi))*(cos(m*theta1-i*sin(theta1)))
            xn1 += y1
        for k in range(1,int(((tramos/2)-1)+1)):
            theta2 = a +(2*k*h)
            y2 = (h/(3*pi))*(cos(m*theta2-i*sin(theta2)))
            xn2 += y2
        xm = 4*xn1+2*xn2
        f_x = (h/3)*(1/pi)*(cos(m*a-i*sin(a))+cos(m*b-i*sin(b)))+xm
        print("Para X= ",i,"; f(x)= ",f_x)

function_x_theta(tramos,a,h)
```

Grafica de las 3 Funciones de Bessel para m=0,m=1,m=2

```python
from numpy import cos,sin,pi
from pylab import *

N = 1000
a = 0
b = pi
h = (b-a)/N

def J(m,x):

    def f(teta):
        return cos(m*teta - x*sin(teta))
    
    s = f(a) + f(b) + 4*f(b-h)
    for k in range(1,N//2):
        s += 4*f(a + (2*k-1)*h) + 2*f(a+2*k*h)
        
    I = h/3*s/pi
    
    return I

x = linspace(0,20)
plot(x,J(0,x),label='J0')
plot(x,J(1,x),label='J1')
plot(x,J(2,x),label='J2')
legend()
```
